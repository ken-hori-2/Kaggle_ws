はい、Pythonを使用して箱ひげ図を作成し、中央値を取得する方法を示します。以下は、pandasとmatplotlibを使用したサンプルコードです：

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# サンプルデータの作成
data = {
    'MR': ['A', 'A', 'B', 'B', 'C', 'C', 'A', 'B', 'C'],
    'KPL': [10, 15, 20, 25, 30, 35, 40, 45, 50]
}
df = pd.DataFrame(data)

# 箱ひげ図の作成
plt.figure(figsize=(10, 6))
sns.boxplot(x='MR', y='KPL', data=df)
plt.title('箱ひげ図: MRとKPLの関係')
plt.show()

# 各グループの中央値を計算
medians = df.groupby('MR')['KPL'].median()
print("各グループの中央値:")
print(medians)

# 全体の中央値を計算
overall_median = df['KPL'].median()
print(f"\n全体の中央値: {overall_median}")
```

このコードは以下のことを行います：

1. 必要なライブラリをインポートします。
2. サンプルデータを作成します（実際のデータを使用する場合は、この部分を置き換えてください）。
3. 箱ひげ図を作成し表示します。
4. 各MRグループごとのKPLの中央値を計算し表示します。
5. 全体のKPLの中央値を計算し表示します。

このコードを実行すると、箱ひげ図が表示され、その後コンソールに各グループの中央値と全体の中央値が出力されます。これにより、視覚的な表現と具体的な数値の両方を得ることができます。

実際のデータを使用する場合は、`pd.read_csv()`や`pd.read_excel()`などの関数を使用してデータを読み込むことができます。
